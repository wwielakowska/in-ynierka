```python
import pandas as pd
wskazniki = pd.read_csv('wskazniki.csv',sep=";", quotechar='"')
```

```python
print(wskazniki)
```

```python
for i, col in enumerate(wskazniki.columns):
    print(f"{i}: {col}")
    
```

```python

wskazniki_col = wskazniki.melt(id_vars=['Nazwa', 'Kod'],    #zostawiamy
                  var_name='Okres',     #Nowa Kolumna
                  value_name='Wskaźnik')  #Kolumna dla wartości

column_parts = wskazniki_col['Okres'].str.split(';', expand=True)
column_parts
```

```python
wskazniki_col['Kwartał'] = column_parts[0] 
wskazniki_col['Przestępstwo'] = column_parts[2]
wskazniki_col['Rok'] = column_parts[3]
wskazniki_col['Województwo'] = wskazniki_col['Nazwa']

# Final DataFrame
wskazniki = wskazniki_col[['Województwo', 'Przestępstwo', 'Kwartał', 'Rok', 'Wskaźnik']]

# Sorting
wskazniki = wskazniki.sort_values(['Rok', 'Kwartał'])


wskazniki.head(15)
```

```python
wskazniki.info()
```

```python
wskazniki["Rok"] = wskazniki["Rok"].astype(int)
wskazniki.info()
```

```python
wskazniki = wskazniki[ ~wskazniki['Przestępstwo'].str.contains("razem", case=False, na=False)]
wskazniki['Przestępstwo'].unique()
```

```python
#Wyodrębienie okresów czasu, aby obliczyć wszystkie kwartały

Q1 = wskazniki[wskazniki["Kwartał"] == "1 kwartał"]
Q12 = wskazniki[wskazniki["Kwartał"] == "pierwsze półrocze"]
Q13 =  wskazniki[wskazniki["Kwartał"] == "1-3 kwartały"]
Y =  wskazniki[wskazniki["Kwartał"] == "rok"]

#Kwartał 2
Q2 = pd.merge(
    Q1, Q12,
    on=["Województwo", "Przestępstwo", "Rok"],
    suffixes=("_Q1", "_Q12")
)

Q2["Wskaźnik"] = Q2["Wskaźnik_Q12"] - Q2["Wskaźnik_Q1"]
Q2["Kwartał"] = "2 kwartał"
Q2= Q2.drop(["Kwartał_Q1", "Wskaźnik %_Q1", "Kwartał_Q12", "Wskaźnik %_Q12"], axis=1)




#Kwartał 3
Q3 = pd.merge(
    Q12, Q13,
    on=["Województwo", "Przestępstwo", "Rok"],
    suffixes=("_Q12", "_Q13")
)

Q3["Wskaźnik %"] = Q3["Wskaźnik %_Q13"] - Q3["Wskaźnik %_Q12"]
Q3["Kwartał"] = "3 kwartał"
Q3= Q3.drop(["Kwartał_Q12", "Wskaźnik %_Q12", "Kwartał_Q13", "Wskaźnik %_Q13"], axis=1)


#Kwartał 4
Q4 = pd.merge(
    Q13, Y,
    on=["Województwo", "Przestępstwo", "Rok"],
    suffixes=("_Q13", "_Y")
)

Q4["Wskaźnik %"] = Q4["Wskaźnik %_Y"] - Q4["Wskaźnik %_Q13"]
Q4["Kwartał"] = "4 kwartał"
Q4= Q4.drop(["Kwartał_Q13", "Wskaźnik %_Q13", "Kwartał_Y", "Wskaźnik %_Y"], axis=1)

wskazniki = pd.concat([wskazniki, Q2, Q3, Q4], ignore_index=True)
wskazniki
```
