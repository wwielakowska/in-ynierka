```python
import pandas as pd
ludnosc = pd.read_csv('ludnosc.csv',sep=";", quotechar='"')
ludnosc = ludnosc.drop(columns='Kod')
```

```python
for i, col in enumerate(ludnosc.columns):
    print(f"{i}: {col}")
```

```python
ludnosc_long = ludnosc.melt(id_vars=['Nazwa'],    #zostawiamy
                  var_name='Okres czasu',     #Nowa Kolumna
                  value_name='Ludność')  #Kolumna dla wartości
ludnosc_long
```

```python
print(ludnosc_long['Ludność'].apply(type).value_counts())
```

```python
column_parts = ludnosc_long['Okres czasu'].str.split(';', expand=True)
column_parts
```

```python
ludnosc_long['Okres czasu'] = column_parts[0]
ludnosc_long['Rok'] = column_parts[3]
ludnosc_long['Województwo'] = ludnosc_long['Nazwa']

# Final DataFrame
df_ludnosc = ludnosc_long[['Województwo', 'Okres czasu', 'Rok','Ludność']]

# Sorting
df_ludnosc = df_ludnosc.sort_values(['Rok'])
```

```python
df_ludnosc = df_ludnosc.dropna()
```

```python
df_ludnosc['Ludność_100tys'] = df_ludnosc['Ludność']/100000
df_ludnosc
```

```python
df_ludnosc.to_csv('ludnosc', index=False)
```
