```python
hist = df_ludnosc_hist[df_ludnosc_hist['Województwo'] != 'POLSKA']
new = df_ludnosc[df_ludnosc['Województwo'] != "POLSKA"]

results = {}
seasonal = hist.groupby(['Województwo', "Okres czasu"])["Ludność"].mean()

for woj in hist['Województwo'].unique():
    df = new[new['Województwo']==woj].set_index(['Rok', 'Okres czasu'])['Ludność'].unstack()
    
    q_means = seasonal.loc[woj].values
    q4 = q_means[-1]             
    a = q_means / q4 
    
    quarterly_list = []
    years = df.index
    for yr in years:
        h1 = df.loc[yr,1]
        h2 = df.loc[yr,2]
        q1 = h2 * a[0]        
        q2 = h1
        q3 = h2 * a[2]
        q4 = h2
        quarterly_list.extend([q1, q2, q3, q4])
    
    # indeks kwartalny
    idx = pd.period_range(start=str(years[0])+'Q1', periods=len(quarterly_list), freq='Q')
    results[woj] = pd.Series(quarterly_list, index=idx)

# łączymy wszystkie województwa
quarterly_df = pd.concat(results, names=['Województwo','Period'])
quarterly_df = quarterly_df.rename("Ludność").reset_index()
quarterly_df['Rok'] = quarterly_df['Period'].dt.year
quarterly_df['Kwartał'] = quarterly_df['Period'].dt.quarter
quarterly_df['Ludność'] = quarterly_df['Ludność'].astype(int)


#adding poland as sum of regions
poland = quarterly_df.groupby('Period')['Ludność'].sum().reset_index()
poland['Województwo'] = 'POLSKA'
poland['Rok'] = poland['Period'].dt.year
poland['Kwartał'] = poland['Period'].dt.quarter


ludnosc_df = pd.concat([quarterly_df, poland], ignore_index=True)
annual = ludnosc_df[ludnosc_df['Kwartał']==4].copy()
annual['Kwartał'] = 0
ludnosc_df = pd.concat([ludnosc_df, annual], ignore_index=True)

cols = ['Województwo', 'Kwartał', 'Rok', 'Ludność']
ludnosc_df = ludnosc_df[cols]

ludnosc_df['Ludność_100tys'] = ludnosc_df['Ludność']/100000
ludnosc_df.to_csv('ludnosc.csv', index=False)
ludnosc_df
```
