```python
import pandas as pd
rozwody = pd.read_csv('rozwody.csv',sep=";", quotechar='"')
rozwody = rozwody.drop(columns='Kod')

for i, col in enumerate(rozwody.columns):
    print(f"{i}: {col}")
```

```python
rozwody_long = rozwody.melt(id_vars=['Nazwa'],    #zostawiamy
                  var_name='Kwartał',     #Nowa Kolumna
                  value_name='Rozwody')  #Kolumna dla wartości
rozwody_long
```

```python
column_parts = rozwody_long['Kwartał'].str.split(';', expand=True)
column_parts
```

```python
rozwody_long['Kwartał'] = column_parts[0]
rozwody_long['Rok'] = column_parts[2]
rozwody_long['Województwo'] = rozwody_long['Nazwa']

# Final DataFrame
df_rozwody = rozwody_long[['Województwo', 'Kwartał', 'Rok','Rozwody']]

# Sorting
df_rozwody = df_rozwody.sort_values(['Rok'])
```

```python
df_rozwody = df_rozwody[ ~df_rozwody['Kwartał'].str.contains("pierwsze półrocze", case=False, na=False)]
df_rozwody['Kwartał'].unique()
```

```python
mapa_kwartał = {
    "1 kwartał": 1,
    "2 kwartał": 2,
    "3 kwartał": 3,
    "4 kwartał": 4,
    "rok": 0
}

df_rozwody["Kwartał"] = df_rozwody["Kwartał"].map(mapa_kwartał)
```

```python
df_rozwody = df_rozwody.dropna()
```
