```python
import pandas as pd
bezrobocie = pd.read_csv('bezrobocie.csv',sep=";", quotechar='"')
bezrobocie = bezrobocie.drop(columns='Kod')
```

```python
for i, col in enumerate(bezrobocie.columns):
    print(f"{i}: {col}")
```

```python
bezrobocie = bezrobocie.loc[:, bezrobocie.columns.str.contains("stopa bezrobocia rejestrowanego", case=False)| 
                        bezrobocie.columns.str.contains("Nazwa", case=False)]

for i, col in enumerate(bezrobocie.columns):
    print(f"{i}: {col}")
```

```python
bezrobocie_long = bezrobocie.melt(id_vars=['Nazwa'],    #zostawiamy
                  var_name='Miesiąc',     #Nowa Kolumna
                  value_name='Stopa bezrobocia [%]')  #Kolumna dla wartości
bezrobocie_long
```

```python
print(bezrobocie_long['Stopa bezrobocia [%]'].apply(type).value_counts())
```

```python
column_parts = bezrobocie_long['Miesiąc'].str.split(';', expand = True)
column_parts
```

```python
bezrobocie_long['Miesiąc'] = column_parts[0]
bezrobocie_long['Rok'] = column_parts[2]
bezrobocie_long['Województwo'] = bezrobocie_long['Nazwa']

df_bezrobocie = bezrobocie_long[['Województwo', 'Miesiąc', 'Rok', 'Stopa bezrobocia [%]']]
```

```python
df_bezrobocie = df_bezrobocie.sort_values(['Rok'])
df_bezrobocie = df_bezrobocie.dropna()
df_bezrobocie
```

### Obliczanie kwartałów

```python
miesiac_map = {
    "styczeń": "1", "luty": '1', "marzec": '1',
    "kwiecień": '2', "maj": '2', "czerwiec": '2',
    "lipiec": '3', "sierpień": '3', "wrzesień": '3',
    "październik": '4', "listopad": '4', "grudzień": '4'
}

df_bezrobocie["Kwartał"] = df_bezrobocie['Miesiąc'].map(miesiac_map)
```

```python
df_bezrobocie.info()
```

```python
df_bezrobocie["Kwartał"] = df_bezrobocie["Kwartał"].astype(int)

df_bezrobocie["Stopa bezrobocia [%]"] = df_bezrobocie["Stopa bezrobocia [%]"].str.replace(',','.')
df_bezrobocie["Stopa bezrobocia [%]"] = df_bezrobocie["Stopa bezrobocia [%]"].astype(float)
```

```python
df_bezrobocie
```

```python
df_bezrobocie_kw = (
    df_bezrobocie.groupby(['Województwo','Kwartał','Rok'], as_index=False)['Stopa bezrobocia [%]']
      .mean()
)
```

```python
df_bezrobocie_kw = df_bezrobocie_kw.sort_values(['Rok','Kwartał'])
df_bezrobocie_kw
```

```python
df_bezrobocie_r = (
    df_bezrobocie_kw.groupby(['Województwo','Rok'], as_index=False)['Stopa bezrobocia [%]']
      .mean()
)
df_bezrobocie_r["Kwartał"] = 0
df_bezrobocie_r
```

```python
df_bezrobocie_final = pd.concat([df_bezrobocie_kw, df_bezrobocie_r], ignore_index=True)
df_bezrobocie_final
```

```python
df_bezrobocie_final.isna().sum()
```

```python
df_bezrobocie_final.to_csv('stopa_bezrobocia', index=False)
```
