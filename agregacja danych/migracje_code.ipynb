# Migracje 2016-2024

```python
import pandas as pd
migracje = pd.read_csv('migracje_GUS.csv',sep=";", quotechar='"')
migracje = migracje.drop(columns='Kod')

for i, col in enumerate(migracje.columns):
    print(f"{i}: {col}")
```

```python
df_migracje.isna().sum()
```

```python
mapa_kwartał = {
    "pierwsze półrocze": 1,
    "rok": 2
}

df_migracje["Okres czasu"] = df_migracje["Okres czasu"].map(mapa_kwartał)
df_migracje["Okres czasu"] = df_migracje["Okres czasu"].astype(int)
df_migracje["Rok"] = df_migracje["Rok"].astype(int)
```

# Agregacja

```python
from statsmodels.tsa.interp.denton import dentonm

for typ in df_migracje['Rodzaj migracji'].unique():
    hist = stare_mig[stare_mig["Rodzaj migracji"]==typ]
    new = df_migracje[(df_migracje['Rodzaj migracji'] == typ)&(df_migracje['Okres czasu']==2)]
    results = {}
    
    #we calculate seasonal indicators:
    seasonal = (
    hist
    .groupby(['Województwo',"Okres czasu"])["Saldo migracji"]
    .mean()
    )

    seasonal = seasonal / seasonal.groupby('Województwo').sum()
    
    for woj in hist['Województwo'].unique():
        # annual values for each region with year as index
        annual = new[new['Województwo'] == woj]['Saldo migracji']
        annual.index = pd.period_range(start=str(2016), periods=len(annual), freq='A')

        # qutarelry indicators for region
        shares = seasonal.loc[woj].values

        # quarterly indicators in list - for each year we have the same quarterly indicators
        indicator = shares.tolist() * len(annual)

        # denton
        y_high = dentonm(
            benchmark=annual.values,
            indicator=indicator,
            h=4,
            conversion='sum',
            method='ratio',
            additive=False
        )

        # quarterly index
        quarterly_index = pd.period_range(start=str(annual.index[0].year)+'Q1', periods=len(y_high), freq='Q')
        quarterly_series = pd.Series(y_high, index=quarterly_index)
        results[woj] = quarterly_series
    
    quarterly = pd.concat(results, names=['Województwo', 'Period'])
    quarterly = quarterly.rename("Saldo migracji").reset_index()


    quarterly = quarterly.fillna(0)
    quarterly['Saldo migracji'] = quarterly['Saldo migracji'].astype(int)
    quarterly['Rok'] = quarterly['Period'].dt.year   
    quarterly['Kwartał'] = quarterly['Period'].dt.quarter
    
    
    annual_df = new[['Województwo', 'Rok', 'Saldo migracji']]
    annual_df['Kwartał'] = 0
    
    migracje_df = pd.concat(
    [
        quarterly[['Województwo','Rok','Kwartał','Saldo migracji']],
        annual_df[['Województwo','Rok','Kwartał','Saldo migracji']]
    ],
    ignore_index=True
)
    
    migracje_df[f"{typ}"] = migracje_df['Saldo migracji']
    migracje_df = migracje_df.drop(columns='Saldo migracji')
    
    
    migracje_df.to_csv(f'{typ}.csv', index=False)
```
