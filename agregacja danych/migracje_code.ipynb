```python
import pandas as pd
migracje = pd.read_csv('migracje.csv',sep=";", quotechar='"')
migracje = migracje.drop(columns='Kod')

for i, col in enumerate(migracje.columns):
    print(f"{i}: {col}")
```

```python
migracje = migracje.loc[:, migracje.columns.str.contains("saldo migracji", case=False)| 
                        migracje.columns.str.contains("Nazwa", case=False)]

for i, col in enumerate(migracje.columns):
    print(f"{i}: {col}")
```

```python
migracje_long = migracje.melt(id_vars=['Nazwa'],    #zostawiamy
                  var_name='Okres czasu',     #Nowa Kolumna
                  value_name='Saldo migracji')  #Kolumna dla wartości
migracje_long
```

```python
column_parts = migracje_long['Okres czasu'].str.split(';', expand=True)
column_parts
```

```python
migracje_long['Okres czasu'] = column_parts[0]
migracje_long['Rok'] = column_parts[2]
migracje_long['Województwo'] = migracje_long['Nazwa']
migracje_long['Rodzaj migracji'] = column_parts[1]

# Final DataFrame
df_migracje = migracje_long[['Województwo', 'Okres czasu', 'Rok','Rodzaj migracji','Saldo migracji']]

# Sorting
df_migracje = df_migracje.sort_values(['Rok'])
```

```python
df_migracje.isna().sum()
```

```python
df_migracje
```

```python
df_migracje['Rodzaj migracji'] = (df_migracje['Rodzaj migracji']
                                  .str.replace('saldo migracji wewnętrznych', 'wewnętrzne')
                                  .str.replace('saldo migracji zagranicznych', 'zagraniczne'))
                                  
```
