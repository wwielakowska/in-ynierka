```python
import pandas as pd
ludnosc = pd.read_csv('ludnosc.csv',sep=";", quotechar='"')
ludnosc = ludnosc.drop(columns='Kod')
```

```python
for i, col in enumerate(ludnosc.columns):
    print(f"{i}: {col}")
```

```python
ludnosc_long = ludnosc.melt(id_vars=['Nazwa'],    #zostawiamy
                  var_name='Rok',     #Nowa Kolumna
                  value_name='Ludnosc w tys')  #Kolumna dla wartości
ludnosc_long
```

```python
column_parts = ludnosc_long['Rok'].str.split(';', expand=True)
column_parts
```

```python
ludnosc_long['Rok'] = column_parts[1]
ludnosc_long['Województwo'] = ludnosc_long['Nazwa']

# Final DataFrame
df_ludnosc = ludnosc_long[['Województwo', 'Rok', 'Ludnosc w tys']]

# Sorting
df_ludnosc = df_ludnosc.sort_values(['Rok'])
```

```python
df_ludnosc = df_ludnosc.dropna()
```

```python
df_ludnosc['Ludnosc w tys'] = df_ludnosc['Ludnosc w tys'].str.replace(',','.')
df_ludnosc['Ludnosc w tys'] = df_ludnosc['Ludnosc w tys'].astype(float)
```

```python
df_ludnosc['Lud_100tys'] = df_ludnosc['Ludnosc w tys']/100
df_ludnosc = df_ludnosc.drop(columns= 'Ludnosc w tys')
```
